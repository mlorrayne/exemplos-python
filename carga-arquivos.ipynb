```python
#Importar bibliotecas
import csv
import pandas as pd
```

```python
#Abrir arquivo (informação copiada da Wikipedia: https://pt.wikipedia.org/wiki/Machado_de_Assis)

arquivo=open('machado_de_assis.txt','r')
```

```python
#Ler linhas, fechar o arquivo e mostrar conteúdo:

conteudo = arquivo.readlines()
arquivo.close()
```

```python
print(conteudo)
print('-----------')
for linha in conteudo:
    print(linha)
```

```python
#Alterar em todo o texto a palavra "Rio de Janeiro" por "Ribeirão Preto"

x=''
for linha in conteudo:
    if linha != '':
        x+=linha

print(x)
x=x.replace('Rio de Janeiro', 'Ribeirão Preto')
print('------')
print(x)
```

```python
#Gerar um novo arquivo

arquivo_novo = open('machado_de_assis2.txt', 'w')
arquivo_novo.writelines(x)
arquivo_novo.close()
```

#Trabalhando com arquivos CSV

Abrindo arquivo CSV - Cotação das ações da Tesla
https://www.kaggle.com/rpaguirre/tesla-stock-price?select=Tesla.csv+-+Tesla.csv.csv


```python
import csv
with open('kaggle_tesla.csv', 'r') as f:
    leitor = csv.reader(f, delimiter=',')
    for linha in leitor:
        data = linha[0]
        fechamento = linha[4]
        volume = linha[5]
        print('Data: ', data, '--Fechamento: ', fechamento, '--Volume: ', volume)
```
